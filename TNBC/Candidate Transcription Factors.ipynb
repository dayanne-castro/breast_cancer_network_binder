```python
metadata = pd.read_csv('./BRCA.1218_pam50.FINAL.txt', sep = '\t')
metadata.index = metadata['Barcode']
metadata = metadata.loc[activities.columns,:]
```

```python
act = StandardScaler().fit_transform(activities.transpose())
act = pd.DataFrame(act, columns = activities.index)
act.loc[:,'sample'] = metadata['Call'].tolist()
df = pd.melt(act, id_vars = 'sample')
df = df.sort_values('sample')

BASALvsNORMAL_act = []

for regulator in df['variable'].unique():
    df_tmp = df.loc[df['variable'] == regulator,:]
    BASAL = df_tmp.loc[df_tmp['sample'] == 'Basal','value']
    NORMAL = df_tmp.loc[df_tmp['sample'] == 'Normal','value']
    BASALvsNORMAL_act.append([regulator, BASAL.mean()-NORMAL.mean(), stats.ttest_ind(BASAL, NORMAL, equal_var = False)[1]])
        
BASALvsNORMAL_act = pd.DataFrame(BASALvsNORMAL_act, columns = ['gene', 'change', 'p_value'])
BASALvsNORMAL_act['fdr'] = multipletests(BASALvsNORMAL_act['p_value'], alpha = 0.05, method = 'fdr_bh')[1]
```

```python
BASALvsNORMAL_act = BASALvsNORMAL_act.sort_values('fdr')
BASALvsNORMAL_act.index = BASALvsNORMAL_act['gene']
del BASALvsNORMAL_act.index.name
```

```python
deseq = pd.read_csv('Basal_DE_FCgreater1.tsv', sep = ' ')
genes_basal_up = deseq.loc[(deseq['padj'] < 0.01)&(deseq['log2FoldChange'] > 0),:].index
genes_basal_down = deseq.loc[(deseq['padj'] < 0.01)&(deseq['log2FoldChange'] < 0),:].index
```

```python
def get_network_dict(network):
    
    regulators = network['regulator'].unique()

    network_dict = {}
    
    for regulator in regulators:
        net_tmp = network.loc[network['regulator'] == regulator, :]
        targets = net_tmp['target']
        boots_perc = net_tmp['bootstraps'].abs()/network['bootstraps'].max()
        #ind_regulators = targets[targets.isin(regulators)]
        ind_targets = []
        #for ind_reg in ind_regulators:
        #    ind_targets.append(network.loc[network['regulator'] == ind_reg, 'target'].tolist())
        #ind_targets = sum(ind_targets, [])    
        network_dict[regulator] = (sum([targets.tolist(), ind_targets], []), boots_perc)
    
    return(network_dict)


def rank_regulators(network, genesDE, genes, target_min = 3, regulators = None):
    
    M = len(genes)
    n = len(genesDE)
    
    network_dict = get_network_dict(network)
    
    df = []
    
    if regulators is None:
        regulators = network_dict.keys()
    else:
        regulators = [reg for reg in network_dict.keys() if reg in regulators]
    
    for regulator in regulators:
        k = len([target for target in network_dict[regulator][0] if target in genesDE])        
        targets_bool = np.asarray([target in genesDE for target in network_dict[regulator][0]])
        #k = np.sum(np.asarray(network_dict[regulator][1])[targets_bool])
        N = len(network_dict[regulator][0])
        #N = network['regulator'].value_counts().mean()
        if N >= target_min:
            pvalue = hypergeom.sf(k, M, n, N)
            score = -np.log10(pvalue)
            df.append((regulator, score, pvalue))
    
    if len(df) > 0:
        df = pd.DataFrame(df, columns = ['regulator', 'score', 'p_value'])
        df['fdr'] = multipletests(df['p_value'], alpha = 0.05, method = 'fdr_bh')[1]
        df = df.sort_values('fdr')
    
    return(df) 
```
